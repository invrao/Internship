# Q1.Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#connecting to the webdriver
driver=webdriver.Edge(r"C:\Users\invra\Downloads\edgedriver_win64\msedgedriver.exe")
```

```python
#maximizing the window
driver.maximize_window()
```

```python
#opening the given wikipedia link on automated edge browser
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

time.sleep(3)
```

```python
#Creating empty lists for saving the required scraping details

Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]

#getting the rank of video
try:
    Ranks=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in Ranks:
        Rank.append(i.text.replace('.', '').replace(' ', ''))
except NoSuchElementException:
    Rank.append("-")    

    
#getting the Name of the video
try:
    Names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in Names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-") 
    

#getting the Name of the artist for the video
try:
    Artists=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in Artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-") 

    

#getting the views in billion for the video
try:
    View=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in View:
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-") 
    

#getting the uploaded date of the video
try:
    Upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')    
    for i in Upload:
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append("-") 
    
```

```python
#Creating a pandas data frame
Top_YT_Videos=pd.DataFrame({"Rank":Rank,"Video_name":Name,"Artist":Artist,"Publication date":Upload_date,"Views (billions)":Views})
Top_YT_Videos
```

```python
# removing the last numbers from video_name column
Top_YT_Videos.Video_name = Top_YT_Videos.Video_name.apply(lambda x:x[:-4].strip('"'))
Top_YT_Videos
```

```python
driver.close()
```

# Q5.Scrape the details of trending repositories on Github.com.

D) Language used


```python
#importing required libraries
import selenium
import pandas as pd
import time
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import requests
import re
```

```python
#opening the given github link on automated edge browser
driver.get("https://github.com/")

time.sleep(3)
```

```python
# getting explore button and clicking on it
explore = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]").click()

# selecting trending option
trend_url = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a")
urls = trend_url.get_attribute("href")
driver.get(urls)
```

```python
#creating empty lists for saving the details

Title = []
Desc = []
Contr_Count = []
Language = []
URL = []
lang=[]


#As all the details are in the individual repository pages
# fetching urls for each repository
repo = driver.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']//a")
for i in repo:
    URL.append(i.get_attribute("href"))
    

# for getting Repository title data
titles = driver.find_elements(By.XPATH,"//h1[@class = 'h3 lh-condensed']")
for i in titles:
    Title.append(i.text)
    
# scraping individual data from all repository pages
for i in URL:
    driver.get(i)
    time.sleep(5)
    
    # Getting Repository Description data 
    try:
        Description = driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Desc.append(Description.text)
    except NoSuchElementException:
        Desc.append('NA')
        
        
    # scraping Contributors Count data
    try:
        contributors = driver.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        Contr_Count.append(contributors.text.replace('Contributors',''))
    except NoSuchElementException:
        Contr_Count.append('NA')

        
    # scraping Languages used data
    try:
        #lang = driver.find_elements(By.XPATH,"//ul[@class= 'list-style-none']//li//span[1]")
        for i in driver.find_elements(By.XPATH,"//ul[@class= 'list-style-none']//li//span[1]"):
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('NA')
        
```

```python
#Creating pandas Data Frame

Repositories = pd.DataFrame({"Repo_Title":Title,"Description":Desc,"Contributors Count":Contr_Count})
Repositories
```

```python
# creating empty lists
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attributes = []
Year = []

for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    # scraping  Dataset name
    try:
        dataset_name = driver.find_element(By.XPATH,"//span[@class='heading']")
        Dataset_name.append(dataset_name.text)
    except NoSuchElementException:
        Dataset_name.append('-')
   
    
    
    
    # scraping data type
    try:
        data_type = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr/td[2]")
        if data_type.text == "N/A": raise NoSuchElementException
        Data_type.append(data_type.text)
    except NoSuchElementException:
        Data_type.append('-')
   
    
    
    # scraping Task
    try:
        task = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[3]/td[2]")
        if task.text == "N/A": raise NoSuchElementException
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
   
    
    
    
    # scraping Attribute type
    try:
        attribute_type = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[2]")
        if attribute_type.text == "N/A": raise NoSuchElementException
        Attribute_type.append(attribute_type.text)
    except NoSuchElementException:
        Attribute_type.append('-')
  
    
    
    
    # scraping No of Instances
    try:
        instances = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr/td[4]")
        if instances.text == "N/A": raise NoSuchElementException
        No_of_instances.append(instances.text)
    except NoSuchElementException:
        No_of_instances.append('-')
   
    
    
    
    # scraping No of Arrtibutes
    try:
        attribute = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[4]")
        if attribute.text == "N/A": raise NoSuchElementException
        No_of_attributes.append(attribute.text)
    except NoSuchElementException:
        No_of_attributes.append('-')
  
    
    
    # scraping Year
    try:
        year = driver.find_element(By.XPATH,"//table[@border='1']//tbody/tr[2]/td[6]")
        if year.text == "N/A": raise NoSuchElementException
        Year.append(year.text[:4])
    except NoSuchElementException:
        Year.append('-')
 
    
```

```python
# Interupted manually as it taking long to collect data, can be further collected data, if not interrupted

#creating pandas data frame
UCI_ML=pd.DataFrame({"Dataset_name":Dataset_name,"Data_type":Data_type,"Task":Task,"Attribute_type":Attribute_type,"No_of_instances":No_of_instances,"No_of_attributes":No_of_attributes,"Year":Year})
UCI_ML
```
